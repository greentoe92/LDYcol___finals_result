```python
from scipy.stats import *

jazz = ['LLP', 'INAT', 'BGGN', 'total'] #, 'target']
methods = list(normal_nov.columns)

for j in methods:
    pot = []
    print('<<<<<<<<<'+j+'>>>>>>>>')
    for i in jazz:
        model = stats.linregress(RAM_nov[i], normal_nov[j])
        print(i)
        print("\n model: ", model)
        print("기울기: ", model.slope)
        print("절편: ", model.intercept) 
        print("상관계수: ", model.rvalue) 
        print("p값: ", model.pvalue)
        print("표준오차: ", model.stderr)
        print("결정계수(R2): ", model.rvalue**2)
        
        X = RAM_nov[i]
        X = sm.add_constant(X)
        model2 = sm.OLS(normal_nov[j], X).fit()      
        model2.summary()
        print()
        print()

        if model.pvalue < 0.05:
            pot.append(i)

    pot
```
