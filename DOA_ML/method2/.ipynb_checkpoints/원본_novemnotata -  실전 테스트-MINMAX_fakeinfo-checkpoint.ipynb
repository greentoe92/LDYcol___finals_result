```python
RA_pot_mix = []
Year_pot_mix = []
std_pot_mix = []
var_pot_mix = []
mean_pot_mix = []


# cao.Source[cao[cao.Source == 'INAT'].index] = 'I-naturalist'
# cao.Source[cao[cao.Source == 'LLP'].index] = 'The Lost Ladybugs'
for i in cao_origin.index:
    if cao_origin.Source[i] == 'INAT':
        cao_origin.Source[i] = 'I-naturalist'
    elif cao_origin.Source[i] == 'LLP':
        cao_origin.Source[i] = 'The Lost Ladybugs'

        
for ziz in tqdm(range(0,100)): 
    ### 반복문 시작
    cao = cao_origin.copy()
    
    ############# DEL_sampling 방법
    sample_del = cao[cao.Species == target_species].sample(n=int((quiz*(ratio/100))), random_state = ziz)
    ### sample_del.Year.value_counts() 떨어지는 샘플의 년도 확인
    sample_del_index = sample_del.index
    
    #len(cao.index) ##길이 체크용
    cao.drop(sample_del_index, inplace=True)
    #len(cao.index) ##길이 체크용


    ############# FAKE_sampling 방법
#    len(cao[cao.Species == target_species].index) ##길이 체크용

    sample_fake = cao[cao.Species != target_species].sample(n=int((quiz*(ratio/100))), random_state = ziz*2)

    for i in sample_fake.index:
        cao['Species'][i] = target_species

    # sample_fake.Year.value_counts() # 추가된 샘플의 년도 확인
#    len(cao[cao.Species == target_species].index) ##길이 체크용

    

    ############# RA 구하기

    RAM = pd.DataFrame({"Year":[],"RA":[], "LLP":[], "INAT":[], "total":[], "target":[]})
    RAM['Year'] = cao['Year'].unique()
    RAM = RAM.sort_values(by=['Year']).reset_index(drop=True)


    a=0
    for year in range(2007,2022):
        cake = cao[cao.Year==year]
        numb = len(cake[cake.Species== target_species].index)/len(cake.index)
        RAM['RA'][a] = numb
        a =a+1

    RAM_bipunc = RAM.copy()

    cao_llp = cao[cao.Source == 'The Lost Ladybugs']


    ## LLP
    a=0
    for year in range(2007, 2022):
        cake = cao_llp[cao_llp.Year==year]
        numb = len(cake.index)
        RAM_bipunc['LLP'][a] = numb
        a =a+1


    cao_llp = cao[cao.Source == 'I-naturalist']


    ## INAT
    a=0
    for year in range(2007, 2022):
        cake = cao_llp[cao_llp.Year==year]
        numb = len(cake.index)
        RAM_bipunc['INAT'][a] = numb
        a =a+1


    ## TOTAL
    a=0
    for year in range(2007, 2022):
        cake = cao[cao.Year==year]
        numb = len(cake.index)
        RAM_bipunc['total'][a] = numb
        a =a+1

    ## target observation
    a=0
    for year in range(2007, 2022):
        cake = cao[cao.Year==year]
        cake = cake[cake.Species == target_species]
        numb = len(cake.index)
        RAM_bipunc['target'][a] = numb
        a =a+1


    ############# 감소율 구하기
    frequency = pd.DataFrame({"Start_Year":[],"RA":[]})
    frequency['Start_Year'] = list(range(2007,2013))
    
    try:
        frequency['RA'][0] = (float(RAM_bipunc[RAM_bipunc['Year'] == 2007]['RA']) - float(RAM_bipunc[RAM_bipunc['Year'] == 2016]['RA']))/float(RAM_bipunc[RAM_bipunc['Year'] == 2007]['RA'])
        frequency['RA'][1] = (float(RAM_bipunc[RAM_bipunc['Year'] == 2008]['RA']) - float(RAM_bipunc[RAM_bipunc['Year'] == 2017]['RA']))/float(RAM_bipunc[RAM_bipunc['Year'] == 2008]['RA'])
        frequency['RA'][2] = (float(RAM_bipunc[RAM_bipunc['Year'] == 2009]['RA']) - float(RAM_bipunc[RAM_bipunc['Year'] == 2018]['RA']))/float(RAM_bipunc[RAM_bipunc['Year'] == 2009]['RA'])
        frequency['RA'][3] = (float(RAM_bipunc[RAM_bipunc['Year'] == 2010]['RA']) - float(RAM_bipunc[RAM_bipunc['Year'] == 2019]['RA']))/float(RAM_bipunc[RAM_bipunc['Year'] == 2010]['RA'])
        frequency['RA'][4] = (float(RAM_bipunc[RAM_bipunc['Year'] == 2011]['RA']) - float(RAM_bipunc[RAM_bipunc['Year'] == 2020]['RA']))/float(RAM_bipunc[RAM_bipunc['Year'] == 2011]['RA'])
        frequency['RA'][5] = (float(RAM_bipunc[RAM_bipunc['Year'] == 2012]['RA']) - float(RAM_bipunc[RAM_bipunc['Year'] == 2021]['RA']))/float(RAM_bipunc[RAM_bipunc['Year'] == 2012]['RA'])
    except:
        pass
        

    RA_pot_mix.append(frequency['RA'].max())
    std_pot_mix.append(frequency['RA'].std())
    var_pot_mix.append(frequency['RA'].var())
    mean_pot_mix.append(frequency['RA'].mean())

    
    
    try:
        Year_pot_mix.append(int(frequency[frequency['RA'] == frequency['RA'].max()]['Start_Year'].values))
    except:
        Year_pot_mix.append(0)
                
RA_pot_mix = [0 if math.isnan(x) else x for x in RA_pot_mix]

Year_pot_mix = [0 if math.isnan(x) else x for x in Year_pot_mix]

std_pot_mix = [0 if math.isnan(x) else x for x in std_pot_mix]
var_pot_mix = [0 if math.isnan(x) else x for x in var_pot_mix]
mean_pot_mix = [0 if math.isnan(x) else x for x in mean_pot_mix]

df_reduction['RA'] = RA_pot_mix
df_std['RA'] = std_pot_mix
df_mean['RA'] = mean_pot_mix
df_var['RA'] = var_pot_mix
df_year['RA'] = Year_pot_mix

print('reduction rate')
zza = RA_pot_mix.copy()
zza.sort()
zza

Counter(Year_pot_mix)

print('std')
zza = std_pot_mix.copy()
zza.sort()
zza

print('var')
zza = var_pot_mix.copy()
zza.sort()
zza

print('mean')
zza = mean_pot_mix.copy()
zza.sort()
zza
```

```python
#################### 2007-2016

RA_pot_mix = []
Year_pot_mix = []
reg_box_slope = []
std_pot_mix = []
var_pot_mix = []
mean_pot_mix = []

# cao.Source[cao[cao.Source == 'INAT'].index] = 'I-naturalist'
# cao.Source[cao[cao.Source == 'LLP'].index] = 'The Lost Ladybugs'
for i in cao_origin.index:
    if cao_origin.Source[i] == 'INAT':
        cao_origin.Source[i] = 'I-naturalist'
    elif cao_origin.Source[i] == 'LLP':
        cao_origin.Source[i] = 'The Lost Ladybugs'

        
for ziz in tqdm(range(0,100)):   ################
    ### 반복문 시작
    cao = cao_origin.copy()

    ############# DEL_sampling 방법
    sample_del = cao[cao.Species == target_species].sample(n=int((quiz*(ratio/100))), random_state = ziz)
    ### sample_del.Year.value_counts() 떨어지는 샘플의 년도 확인
    sample_del_index = sample_del.index
    #len(cao.index) ##길이 체크용
    cao.drop(sample_del_index, inplace=True)
    #len(cao.index) ##길이 체크용


    ############# FAKE_sampling 방법
#    len(cao[cao.Species == target_species].index) ##길이 체크용
    sample_fake = cao[cao.Species != target_species].sample(n=int((quiz*(ratio/100))), random_state = ziz*2)
    for i in sample_fake.index:
        cao['Species'][i] = target_species
    # sample_fake.Year.value_counts() # 추가된 샘플의 년도 확인
#    len(cao[cao.Species == target_species].index) ##길이 체크용

    

    ############# RA 구하기

    RAM = pd.DataFrame({"Year":[],"RA":[], "LLP":[], "INAT":[], "total":[], "target":[]})
    RAM['Year'] = cao['Year'].unique()
    RAM = RAM.sort_values(by=['Year']).reset_index(drop=True)


    a=0
    for year in range(2007,2022):
        cake = cao[cao.Year==year]
        numb = len(cake[cake.Species== target_species].index)/len(cake.index)
        RAM['RA'][a] = numb
        a =a+1

    RAM_bipunc = RAM.copy()

    
    ## LLP
    cao_llp = cao[cao.Source == 'The Lost Ladybugs']
    a=0
    for year in range(2007, 2022):
        cake = cao_llp[cao_llp.Year==year]
        numb = len(cake.index)
        RAM_bipunc['LLP'][a] = numb
        a =a+1

## INAT
    cao_llp = cao[cao.Source == 'I-naturalist']
    a=0
    for year in range(2007, 2022):
        cake = cao_llp[cao_llp.Year==year]
        numb = len(cake.index)
        RAM_bipunc['INAT'][a] = numb
        a =a+1


    ## TOTAL
    a=0
    for year in range(2007, 2022):
        cake = cao[cao.Year==year]
        numb = len(cake.index)
        RAM_bipunc['total'][a] = numb
        a =a+1


    ## target observation
    a=0
    for year in range(2007, 2022):
        cake = cao[cao.Year==year]
        cake = cake[cake.Species == target_species]
        numb = len(cake.index)
        RAM_bipunc['target'][a] = numb
        a =a+1


    ############# 감소율 구하기
    frequency = pd.DataFrame({"Start_Year":[],"STE":[]})
    frequency['Start_Year'] = list(range(2007,2013))


    slope = (RAM_bipunc['target'][0]/RAM_bipunc['total'][0] - RAM_bipunc['target'][9]/RAM_bipunc['total'][9])/(2007-(2007+9))
    interc = RAM_bipunc['target'][0]/RAM_bipunc['total'][0] - slope*2007    
    
    for i in range(2007, 2013):
        a = slope*i + interc
        b = slope*(i+9) + interc
        dice = i-2007
        frequency['STE'][dice] = ((a-b)/a)

    RA_pot_mix.append(frequency['STE'].max())
    std_pot_mix.append(frequency['STE'].std())
    var_pot_mix.append(frequency['STE'].var())
    mean_pot_mix.append(frequency['STE'].mean())
        
    
    try:
        Year_pot_mix.append(int(frequency[frequency['STE'] == frequency['STE'].max()]['Start_Year'].values))
    except:
        Year_pot_mix.append(0)
                
RA_pot_mix = [0 if math.isnan(x) else x for x in RA_pot_mix]

Year_pot_mix = [0 if math.isnan(x) else x for x in Year_pot_mix]

std_pot_mix = [0 if math.isnan(x) else x for x in std_pot_mix]
var_pot_mix = [0 if math.isnan(x) else x for x in var_pot_mix]
mean_pot_mix = [0 if math.isnan(x) else x for x in mean_pot_mix]


df_reduction['STE07'] = RA_pot_mix
df_std['STE07'] = std_pot_mix
df_mean['STE07'] = mean_pot_mix
df_var['STE07'] = var_pot_mix
df_year['STE07'] = Year_pot_mix


print('reduction rate')
zza = RA_pot_mix.copy()
zza.sort()
zza

Counter(Year_pot_mix)

print('std')
zza = std_pot_mix.copy()
zza.sort()
zza

print('var')
zza = var_pot_mix.copy()
zza.sort()
zza

print('mean')
zza = mean_pot_mix.copy()
zza.sort()
zza
```
