```python
CV_nov=pd.read_csv("./reduction_rate_range(output)/Coccinella novemnotata_10_STDofABS.csv", index_col=0)
CV_trans=pd.read_csv("./reduction_rate_range(output)/Coccinella transversoguttata_10_STDofABS.csv", index_col=0)
CV_bi=pd.read_csv("./reduction_rate_range(output)/Adalia bipunctata_10_STDofABS.csv", index_col=0)
CV_paren=pd.read_csv("./reduction_rate_range(output)/Hippodamia parenthesis_10_STDofABS.csv", index_col=0)

for i in list(CV_nov.index):
    
    
    #set seaborn plotting aesthetics as default
    sns.set(rc = {'figure.figsize':(12,4)})
    #set_ylabel('Variation in reduction rate')
    sns.set_style(style='white')

    #define plotting region (1 rows, 4 columns)
    fig, axes = plt.subplots(1, 4)

    
    data_nov = CV_nov.loc[:, 'RA':'STE07'][str(i):str(i)]*100
    data_trans = CV_trans.loc[:, 'RA':'STE07'][str(i):str(i)]*100
    data_bi = CV_bi.loc[:, 'RA':'STE07'][str(i):str(i)]*100
    data_paren = CV_paren.loc[:, 'RA':'STE07'][str(i):str(i)]*100

    #create boxplot in each subplot
    a = sns.barplot(data=data_nov, ax=axes[0], palette = "Blues")
    a.set_title('C. novemnotata', fontsize = 13, fontweight='bold')
    a.set(ylabel='std in ' + str(i) + ' rates (%)')
    b= sns.barplot(data=data_trans,ax=axes[1], palette = "Blues")
    b.set_title('C. transversoguttata', fontsize = 13, fontweight='bold')
    c = sns.barplot(data=data_bi, ax=axes[2], palette = "Blues")
    c.set_title('A. bipunctata', fontsize = 13, fontweight='bold')
    d=sns.barplot(data=data_paren, ax=axes[3], palette = "Blues")
    d.set_title('H. parenthesis', fontsize = 13, fontweight='bold')

    plt.show()
    figure = fig.get_figure()    
    figure.savefig('./graph(output)/one_for_all_var_in ' + str(i) +' .pdf', dpi=400)
    
#     figure = fig.get_figure()    
#     figure.savefig('./graph(output)/one_for_all_var_in_reduction.pdf', dpi=400)
```
