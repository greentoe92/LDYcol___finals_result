# 학습용 파일 minmax 정규화

```python
df = df.iloc[:,7:]
```

```python
df.iloc[:,:8]
```

```python
df = df.drop('c9_500m',axis=1)
```

```python
# 피처만 있는 칼럼
df_max = df.iloc[:,8:]
```

```python
df_max['Year']= df['Year']
```

```python
for i in range(2007,2022):
    for j in df_max.columns:
        if df_max.loc[df_max['Year']==i,j].max()!=0 and j != 'Year':
            print(i,j,df_max.loc[df_max['Year']==i,j].max())
            df_max.loc[df_max['Year']==i,j] =  df_max.loc[df_max['Year']==i,j]/df_max.loc[df_max['Year']==i,j].max()
           
```

```python
pd.concat([df.iloc[:,:8],df_max.drop('Year',axis=1)],axis=1).to_csv("./220220_DOA_bi(최종,minmax).csv")
```
